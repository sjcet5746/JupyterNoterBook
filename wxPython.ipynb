```python
#2.Calculator
import wx

class CalculatorFrame(wx.Frame):
    def __init__(self, parent, title):
        super(CalculatorFrame, self).__init__(parent, title=title, size=(300, 400))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.display = wx.TextCtrl(self, style=wx.TE_RIGHT)
        vbox.Add(self.display, flag=wx.EXPAND | wx.ALL, border=10)

        grid = wx.GridSizer(5, 4, 5, 5)

        button_labels = [
            '7', '8', '9', '/',
            '4', '5', '6', '*',
            '1', '2', '3', '-',
            '0', '.', '=', '+',
            'C', 'CE'
        ]

        self.buttons = {}
        for label in button_labels:
            button = wx.Button(self, label=label)
            self.buttons[label] = button
            grid.Add(button, 1, wx.EXPAND)
            button.Bind(wx.EVT_BUTTON, self.OnButtonClick)

        vbox.Add(grid, proportion=1, flag=wx.EXPAND)
        self.SetSizer(vbox)

    def OnButtonClick(self, event):
        label = event.GetEventObject().GetLabel()
        if label == '=':
            try:
                result = str(eval(self.display.GetValue()))
                self.display.SetValue(result)
            except:
                self.display.SetValue("Error")
        elif label == 'C':
            self.display.Clear()
        elif label == 'CE':
            current_text = self.display.GetValue()
            self.display.SetValue(current_text[:-1])
        else:
            current_text = self.display.GetValue()
            new_text = current_text + label
            self.display.SetValue(new_text)

if __name__ == '__main__':
    app = wx.App(False)
    frame = CalculatorFrame(None, "Calculator App")
    app.MainLoop()
```

```python
#3.To Do List
import wx

class ToDoListFrame(wx.Frame):
    def __init__(self, parent, title):
        super(ToDoListFrame, self).__init__(parent, title=title, size=(400, 300))
        
        self.tasks = []
        
        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.task_text = wx.TextCtrl(self)
        add_button = wx.Button(self, label="Add Task")
        self.task_list = wx.ListBox(self, choices=[], style=wx.LB_SINGLE)
        remove_button = wx.Button(self, label="Remove Task")

        add_button.Bind(wx.EVT_BUTTON, self.OnAddTask)
        remove_button.Bind(wx.EVT_BUTTON, self.OnRemoveTask)

        vbox.Add(self.task_text, proportion=0, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(add_button, proportion=0, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.task_list, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(remove_button, proportion=0, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def OnAddTask(self, event):
        task = self.task_text.GetValue()
        if task:
            self.tasks.append(task)
            self.task_list.Set(self.tasks)
            self.task_text.SetValue("")

    def OnRemoveTask(self, event):
        selected_index = self.task_list.GetSelection()
        if selected_index != wx.NOT_FOUND:
            del self.tasks[selected_index]
            self.task_list.Set(self.tasks)

if __name__ == '__main__':
    app = wx.App(False)
    frame = ToDoListFrame(None, "To-Do List App")
    app.MainLoop()
```

```python
#5.Simple Text Editor
import wx
import os

class TextEditorFrame(wx.Frame):
    def __init__(self, parent, title):
        super(TextEditorFrame, self).__init__(parent, title=title, size=(800, 600))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        self.text_ctrl = wx.TextCtrl(self, style=wx.TE_MULTILINE)
        self.status_bar = self.CreateStatusBar()
        self.status_bar.SetStatusText("Ready")

        menubar = wx.MenuBar()
        file_menu = wx.Menu()

        open_file_menu_item = file_menu.Append(wx.ID_OPEN, "Open", "Open a text file")
        save_file_menu_item = file_menu.Append(wx.ID_SAVE, "Save", "Save the current file")
        save_as_file_menu_item = file_menu.Append(wx.ID_SAVEAS, "Save As...", "Save the current file with a different name")
        file_menu.AppendSeparator()
        exit_menu_item = file_menu.Append(wx.ID_EXIT, "Exit", "Exit the application")

        menubar.Append(file_menu, "File")
        self.SetMenuBar(menubar)

        self.Bind(wx.EVT_MENU, self.OnOpen, open_file_menu_item)
        self.Bind(wx.EVT_MENU, self.OnSave, save_file_menu_item)
        self.Bind(wx.EVT_MENU, self.OnSaveAs, save_as_file_menu_item)
        self.Bind(wx.EVT_MENU, self.OnExit, exit_menu_item)

    def OnOpen(self, event):
        wildcard = "Text files (*.txt)|*.txt|All files (*.*)|*.*"
        dialog = wx.FileDialog(self, "Open", "", "", wildcard, wx.FD_OPEN | wx.FD_FILE_MUST_EXIST)
        
        if dialog.ShowModal() == wx.ID_CANCEL:
            return

        file_path = dialog.GetPath()
        with open(file_path, 'r') as file:
            self.text_ctrl.SetValue(file.read())

        self.status_bar.SetStatusText(f"Opened: {file_path}")

    def OnSave(self, event):
        if not hasattr(self, 'current_file'):
            self.OnSaveAs(event)
        else:
            with open(self.current_file, 'w') as file:
                file.write(self.text_ctrl.GetValue())

            self.status_bar.SetStatusText(f"Saved: {self.current_file}")

    def OnSaveAs(self, event):
        wildcard = "Text files (*.txt)|*.txt|All files (*.*)|*.*"
        dialog = wx.FileDialog(self, "Save As", "", "", wildcard, wx.FD_SAVE | wx.FD_OVERWRITE_PROMPT)
        
        if dialog.ShowModal() == wx.ID_CANCEL:
            return

        file_path = dialog.GetPath()
        with open(file_path, 'w') as file:
            file.write(self.text_ctrl.GetValue())

        self.current_file = file_path
        self.status_bar.SetStatusText(f"Saved: {file_path}")

    def OnExit(self, event):
        self.Close()

if __name__ == '__main__':
    app = wx.App(False)
    frame = TextEditorFrame(None, "Simple Text Editor")
    app.MainLoop()
```

```python
#8.Basic Paint App
import wx

class PaintApp(wx.Frame):
    def __init__(self, parent, title):
        super(PaintApp, self).__init__(parent, title=title, size=(800, 600))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.panel.Bind(wx.EVT_PAINT, self.OnPaint)
        self.panel.Bind(wx.EVT_LEFT_DOWN, self.OnMouseDown)
        self.panel.Bind(wx.EVT_LEFT_UP, self.OnMouseUp)
        self.panel.Bind(wx.EVT_MOTION, self.OnMouseMove)

        self.drawing = False
        self.last_x, self.last_y = 0, 0
        self.current_color = wx.Colour(0, 0, 0)

        menubar = wx.MenuBar()
        file_menu = wx.Menu()

        clear_menu_item = file_menu.Append(wx.ID_CLEAR, "Clear", "Clear the canvas")
        file_menu.AppendSeparator()
        exit_menu_item = file_menu.Append(wx.ID_EXIT, "Exit", "Exit the application")

        menubar.Append(file_menu, "File")
        self.SetMenuBar(menubar)

        self.Bind(wx.EVT_MENU, self.OnClear, clear_menu_item)
        self.Bind(wx.EVT_MENU, self.OnExit, exit_menu_item)

    def OnPaint(self, event):
        dc = wx.BufferedPaintDC(self.panel)
        dc.SetBackground(wx.Brush(wx.Colour(255, 255, 255)))
        dc.Clear()
        dc.SetPen(wx.Pen(self.current_color, 5))
        
        if self.drawing:
            dc.DrawLine(self.last_x, self.last_y, self.current_x, self.current_y)

    def OnMouseDown(self, event):
        self.drawing = True
        self.last_x, self.last_y = event.GetPosition()

    def OnMouseUp(self, event):
        self.drawing = False

    def OnMouseMove(self, event):
        if event.Dragging() and self.drawing:
            self.current_x, self.current_y = event.GetPosition()
            self.Refresh()

    def OnClear(self, event):
        self.Refresh()

    def OnExit(self, event):
        self.Close()

if __name__ == '__main__':
    app = wx.App(False)
    frame = PaintApp(None, "Basic Paint App")
    app.MainLoop()
```

```python
#9.Currency Converter
import wx
import requests

class CurrencyConverterFrame(wx.Frame):
    def __init__(self, parent, title):
        super(CurrencyConverterFrame, self).__init__(parent, title=title, size=(400, 200))
        
        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.amount_label = wx.StaticText(self, label="Amount:")
        self.amount_text = wx.TextCtrl(self)
        
        self.from_currency_label = wx.StaticText(self, label="From Currency:")
        self.from_currency_choice = wx.ComboBox(self, choices=["USD", "EUR", "GBP", "JPY"], style=wx.CB_DROPDOWN | wx.CB_READONLY)

        self.to_currency_label = wx.StaticText(self, label="To Currency:")
        self.to_currency_choice = wx.ComboBox(self, choices=["USD", "EUR", "GBP", "JPY"], style=wx.CB_DROPDOWN | wx.CB_READONLY)
        
        self.convert_button = wx.Button(self, label="Convert")
        self.result_label = wx.StaticText(self, label="Result:")
        self.result_text = wx.TextCtrl(self, style=wx.TE_READONLY)

        self.convert_button.Bind(wx.EVT_BUTTON, self.OnConvert)

        vbox.Add(self.amount_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.amount_text, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.from_currency_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.from_currency_choice, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.to_currency_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.to_currency_choice, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.convert_button, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.result_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.result_text, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def OnConvert(self, event):
        amount = self.amount_text.GetValue()
        from_currency = self.from_currency_choice.GetValue()
        to_currency = self.to_currency_choice.GetValue()

        if not amount:
            wx.MessageBox("Please enter an amount to convert.", "Error", wx.OK | wx.ICON_ERROR)
            return

        try:
            amount = float(amount)
        except ValueError:
            wx.MessageBox("Invalid amount. Please enter a valid number.", "Error", wx.OK | wx.ICON_ERROR)
            return

        if from_currency == to_currency:
            result = amount
        else:
            try:
                conversion_rate = self.get_conversion_rate(from_currency, to_currency)
                result = amount * conversion_rate
            except Exception as e:
                wx.MessageBox(f"Error fetching conversion rate: {str(e)}", "Error", wx.OK | wx.ICON_ERROR)
                return

        self.result_text.SetValue(f"{result:.2f} {to_currency}")

    def get_conversion_rate(self, from_currency, to_currency):
        url = f"https://api.exchangerate-api.com/v4/latest/{from_currency}"
        response = requests.get(url)
        data = response.json()
        return data['rates'][to_currency]

if __name__ == '__main__':
    app = wx.App(False)
    frame = CurrencyConverterFrame(None, "Currency Converter App")
    app.MainLoop()
```

```python
#10.BMI Calculator
import wx

class BMICalculatorFrame(wx.Frame):
    def __init__(self, parent, title):
        super(BMICalculatorFrame, self).__init__(parent, title=title, size=(400, 200))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.weight_label = wx.StaticText(self, label="Weight (kg):")
        self.weight_text = wx.TextCtrl(self)
        
        self.height_label = wx.StaticText(self, label="Height (cm):")
        self.height_text = wx.TextCtrl(self)

        self.calculate_button = wx.Button(self, label="Calculate BMI")
        self.result_label = wx.StaticText(self, label="Result:")
        self.result_text = wx.TextCtrl(self, style=wx.TE_READONLY)

        self.calculate_button.Bind(wx.EVT_BUTTON, self.OnCalculateBMI)

        vbox.Add(self.weight_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.weight_text, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.height_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.height_text, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.calculate_button, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.result_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.result_text, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def OnCalculateBMI(self, event):
        weight = self.weight_text.GetValue()
        height = self.height_text.GetValue()

        if not weight or not height:
            wx.MessageBox("Please enter weight and height to calculate BMI.", "Error", wx.OK | wx.ICON_ERROR)
            return

        try:
            weight = float(weight)
            height = float(height) / 100  # Convert height from cm to meters
        except ValueError:
            wx.MessageBox("Invalid input. Please enter valid numbers for weight and height.", "Error", wx.OK | wx.ICON_ERROR)
            return

        bmi = self.calculate_bmi(weight, height)
        self.result_text.SetValue(f"BMI: {bmi:.2f}")

    def calculate_bmi(self, weight, height):
        return weight / (height ** 2)

if __name__ == '__main__':
    app = wx.App(False)
    frame = BMICalculatorFrame(None, "BMI Calculator App")
    app.MainLoop()
```

```python
#11.Stop Watch Timer
import wx
import time

class StopwatchTimerFrame(wx.Frame):
    def __init__(self, parent, title):
        super(StopwatchTimerFrame, self).__init__(parent, title=title, size=(400, 200))

        self.InitUI()
        self.Centre()
        self.Show(True)

        self.running = False
        self.start_time = 0

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.time_display = wx.StaticText(self, label="00:00:00", style=wx.ALIGN_CENTER)
        self.start_stop_button = wx.Button(self, label="Start")
        self.reset_button = wx.Button(self, label="Reset")

        self.start_stop_button.Bind(wx.EVT_BUTTON, self.OnStartStop)
        self.reset_button.Bind(wx.EVT_BUTTON, self.OnReset)

        vbox.Add(self.time_display, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.start_stop_button, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.reset_button, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def OnStartStop(self, event):
        if not self.running:
            self.start_time = time.time() - self.start_time
            self.running = True
            self.start_stop_button.SetLabel("Stop")
            self.timer = wx.Timer(self)
            self.Bind(wx.EVT_TIMER, self.OnUpdateTimer, self.timer)
            self.timer.Start(1000)  # Update the timer display every second
        else:
            self.running = False
            self.start_stop_button.SetLabel("Start")
            self.timer.Stop()

    def OnUpdateTimer(self, event):
        elapsed_time = int(time.time() - self.start_time)
        hours = elapsed_time // 3600
        minutes = (elapsed_time % 3600) // 60
        seconds = elapsed_time % 60
        time_str = f"{hours:02}:{minutes:02}:{seconds:02}"
        self.time_display.SetLabel(time_str)

    def OnReset(self, event):
        self.running = False
        self.start_time = 0
        self.start_stop_button.SetLabel("Start")
        self.time_display.SetLabel("00:00:00")

if __name__ == '__main__':
    app = wx.App(False)
    frame = StopwatchTimerFrame(None, "Stopwatch Timer App")
    app.MainLoop()
```

```python
#15.Basic Drawing Program
import wx

class DrawingFrame(wx.Frame):
    def __init__(self, parent, title):
        super(DrawingFrame, self).__init__(parent, title=title, size=(800, 600))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.canvas = DrawingCanvas(self)
        clear_button = wx.Button(self, label="Clear Canvas")

        clear_button.Bind(wx.EVT_BUTTON, self.OnClearCanvas)

        vbox.Add(self.canvas, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(clear_button, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def OnClearCanvas(self, event):
        self.canvas.ClearDrawing()

class DrawingCanvas(wx.Panel):
    def __init__(self, parent):
        super(DrawingCanvas, self).__init__(parent)

        self.drawing = False
        self.last_x, self.last_y = 0, 0
        self.lines = []

        self.Bind(wx.EVT_LEFT_DOWN, self.OnMouseDown)
        self.Bind(wx.EVT_LEFT_UP, self.OnMouseUp)
        self.Bind(wx.EVT_MOTION, self.OnMouseMove)
        self.Bind(wx.EVT_PAINT, self.OnPaint)

    def OnMouseDown(self, event):
        self.drawing = True
        self.last_x, self.last_y = event.GetPosition()

    def OnMouseUp(self, event):
        self.drawing = False

    def OnMouseMove(self, event):
        if self.drawing:
            x, y = event.GetPosition()
            self.lines.append((self.last_x, self.last_y, x, y))
            self.last_x, self.last_y = x, y
            self.Refresh()

    def OnPaint(self, event):
        dc = wx.PaintDC(self)
        dc.SetPen(wx.Pen(wx.Colour(0, 0, 0), 2))

        for line in self.lines:
            x1, y1, x2, y2 = line
            dc.DrawLine(x1, y1, x2, y2)

    def ClearDrawing(self):
        self.lines = []
        self.Refresh()

if __name__ == '__main__':
    app = wx.App(False)
    frame = DrawingFrame(None, "Basic Drawing Program")
    app.MainLoop()
```

```python
#16.Unit Converter
import wx

class UnitConverterFrame(wx.Frame):
    def __init__(self, parent, title):
        super(UnitConverterFrame, self).__init__(parent, title=title, size=(400, 200))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.value_label = wx.StaticText(self, label="Enter Value:")
        self.value_text = wx.TextCtrl(self)
        
        self.unit_label = wx.StaticText(self, label="Select Unit:")
        self.unit_choice = wx.ComboBox(self, choices=["Feet to Meters", "Miles to Kilometers", "Pounds to Kilograms"], style=wx.CB_DROPDOWN | wx.CB_READONLY)
        
        self.convert_button = wx.Button(self, label="Convert")
        self.result_label = wx.StaticText(self, label="Result:")
        self.result_text = wx.TextCtrl(self, style=wx.TE_READONLY)

        self.convert_button.Bind(wx.EVT_BUTTON, self.OnConvert)

        vbox.Add(self.value_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.value_text, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.unit_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.unit_choice, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.convert_button, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.result_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.result_text, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def OnConvert(self, event):
        value = self.value_text.GetValue()
        conversion_type = self.unit_choice.GetValue()

        if not value:
            wx.MessageBox("Please enter a value to convert.", "Error", wx.OK | wx.ICON_ERROR)
            return

        try:
            value = float(value)
        except ValueError:
            wx.MessageBox("Invalid input. Please enter a valid number.", "Error", wx.OK | wx.ICON_ERROR)
            return

        result = self.perform_conversion(value, conversion_type)
        self.result_text.SetValue(f"{result:.2f}")

    def perform_conversion(self, value, conversion_type):
        if conversion_type == "Feet to Meters":
            return value * 0.3048
        elif conversion_type == "Miles to Kilometers":
            return value * 1.60934
        elif conversion_type == "Pounds to Kilograms":
            return value * 0.453592
        else:
            return value

if __name__ == '__main__':
    app = wx.App(False)
    frame = UnitConverterFrame(None, "Unit Converter App")
    app.MainLoop()
```

```python
#19.Basic Alarm Clock
import wx
import time
import threading

class AlarmClockFrame(wx.Frame):
    def __init__(self, parent, title):
        super(AlarmClockFrame, self).__init__(parent, title=title, size=(400, 200))

        self.InitUI()
        self.Centre()
        self.Show(True)

        self.alarm_thread = None
        self.is_alarm_set = False

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.time_label = wx.StaticText(self, label="Set Alarm Time (HH:MM):")
        self.time_text = wx.TextCtrl(self)
        self.set_button = wx.Button(self, label="Set Alarm")
        self.cancel_button = wx.Button(self, label="Cancel Alarm")

        self.set_button.Bind(wx.EVT_BUTTON, self.OnSetAlarm)
        self.cancel_button.Bind(wx.EVT_BUTTON, self.OnCancelAlarm)

        vbox.Add(self.time_label, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.time_text, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.set_button, flag=wx.EXPAND | wx.ALL, border=10)
        vbox.Add(self.cancel_button, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def OnSetAlarm(self, event):
        alarm_time = self.time_text.GetValue()

        if not alarm_time:
            wx.MessageBox("Please enter a valid alarm time (HH:MM).", "Error", wx.OK | wx.ICON_ERROR)
            return

        try:
            alarm_hour, alarm_minute = map(int, alarm_time.split(":"))
            current_time = time.localtime()
            current_hour, current_minute = current_time.tm_hour, current_time.tm_min

            if (
                alarm_hour < 0
                or alarm_hour > 23
                or alarm_minute < 0
                or alarm_minute > 59
            ):
                raise ValueError

            time_diff = (
                (alarm_hour - current_hour) * 3600 + (alarm_minute - current_minute) * 60
            )

            if time_diff <= 0:
                raise ValueError

            if self.alarm_thread and self.alarm_thread.is_alive():
                self.alarm_thread.join()

            self.alarm_thread = threading.Thread(
                target=self.StartAlarm, args=(time_diff,)
            )
            self.alarm_thread.start()
            self.is_alarm_set = True
            wx.MessageBox(
                f"Alarm set for {alarm_hour:02}:{alarm_minute:02}.",
                "Alarm Set",
                wx.OK | wx.ICON_INFORMATION,
            )
        except ValueError:
            wx.MessageBox(
                "Invalid time format or time has already passed. Please enter a valid future time (HH:MM).",
                "Error",
                wx.OK | wx.ICON_ERROR,
            )

    def OnCancelAlarm(self, event):
        if self.alarm_thread and self.alarm_thread.is_alive():
            self.alarm_thread.join()
        self.is_alarm_set = False
        wx.MessageBox("Alarm canceled.", "Alarm Canceled", wx.OK | wx.ICON_INFORMATION)

    def StartAlarm(self, time_diff):
        time.sleep(time_diff)
        if self.is_alarm_set:
            wx.CallAfter(self.ShowAlarmMessage)

    def ShowAlarmMessage(self):
        wx.MessageBox("Time to wake up!", "Alarm", wx.OK | wx.ICON_INFORMATION)
        self.is_alarm_set = False

if __name__ == "__main__":
    app = wx.App(False)
    frame = AlarmClockFrame(None, "Basic Alarm Clock")
    app.MainLoop()
```

```python
#23.Tic Tac Toe Game
import wx

class TicTacToeFrame(wx.Frame):
    def __init__(self, parent, title):
        super(TicTacToeFrame, self).__init__(parent, title=title, size=(300, 300))

        self.InitUI()
        self.Centre()
        self.Show(True)

        self.board = [[' ' for _ in range(3)] for _ in range(3)]
        self.current_player = 'X'
        self.game_over = False

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.buttons = []
        for i in range(3):
            hbox = wx.BoxSizer(wx.HORIZONTAL)
            for j in range(3):
                button = wx.Button(self, id=wx.ID_ANY, label='', size=(50, 50))
                button.Bind(wx.EVT_BUTTON, self.OnButtonClick)
                hbox.Add(button, flag=wx.EXPAND | wx.ALL, border=1)
                self.buttons.append(button)
            vbox.Add(hbox, proportion=1, flag=wx.EXPAND)
        
        self.SetSizer(vbox)

    def OnButtonClick(self, event):
        if self.game_over:
            return

        button = event.GetEventObject()
        row, col = self.GetButtonPosition(button)

        if self.board[row][col] == ' ':
            button.SetLabel(self.current_player)
            self.board[row][col] = self.current_player
            self.CheckForWin()
            self.TogglePlayer()

    def GetButtonPosition(self, button):
        index = self.buttons.index(button)
        row = index // 3
        col = index % 3
        return row, col

    def TogglePlayer(self):
        self.current_player = 'O' if self.current_player == 'X' else 'X'

    def CheckForWin(self):
        # Check rows
        for row in self.board:
            if row[0] == row[1] == row[2] != ' ':
                self.GameOver(row[0])
                return

        # Check columns
        for col in range(3):
            if self.board[0][col] == self.board[1][col] == self.board[2][col] != ' ':
                self.GameOver(self.board[0][col])
                return

        # Check diagonals
        if self.board[0][0] == self.board[1][1] == self.board[2][2] != ' ':
            self.GameOver(self.board[0][0])
            return
        if self.board[0][2] == self.board[1][1] == self.board[2][0] != ' ':
            self.GameOver(self.board[0][2])
            return

        # Check for a draw
        if all(all(cell != ' ' for cell in row) for row in self.board):
            self.GameOver('Draw')
            return

    def GameOver(self, winner):
        if winner == 'Draw':
            message = "It's a draw!"
        else:
            message = f"Player {winner} wins!"

        wx.MessageBox(message, "Game Over", wx.OK | wx.ICON_INFORMATION)
        self.game_over = True

if __name__ == '__main__':
    app = wx.App(False)
    frame = TicTacToeFrame(None, "Tic-Tac-Toe Game")
    app.MainLoop()
```

```python
#25.Basic Image Slideshow
import wx
import os
import glob
import random

class ImageSlideshowFrame(wx.Frame):
    def __init__(self, parent, title):
        super(ImageSlideshowFrame, self).__init__(parent, title=title, size=(800, 600))

        self.InitUI()
        self.Centre()
        self.Show(True)

        self.image_paths = self.load_images()
        self.current_image_index = 0
        self.timer = wx.Timer(self)
        self.Bind(wx.EVT_TIMER, self.OnNextImage, self.timer)
        self.timer.Start(3000)  # Change image every 3 seconds

    def InitUI(self):
        vbox = wx.BoxSizer(wx.VERTICAL)

        self.image_ctrl = wx.StaticBitmap(self, bitmap=wx.Bitmap(800, 600))
        vbox.Add(self.image_ctrl, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)

        self.SetSizer(vbox)

    def load_images(self):
        image_paths = glob.glob("images/*.jpg")
        random.shuffle(image_paths)
        return image_paths

    def OnNextImage(self, event):
        if self.current_image_index < len(self.image_paths):
            image_path = self.image_paths[self.current_image_index]
            image = wx.Image(image_path, wx.BITMAP_TYPE_ANY)
            image = image.Scale(800, 600, wx.IMAGE_QUALITY_HIGH)
            bitmap = wx.Bitmap(image)
            self.image_ctrl.SetBitmap(bitmap)
            self.current_image_index += 1
        else:
            self.timer.Stop()
            wx.MessageBox("Slideshow finished.", "Slideshow", wx.OK | wx.ICON_INFORMATION)
            self.Close()

if __name__ == '__main__':
    app = wx.App(False)
    frame = ImageSlideshowFrame(None, "Image Slideshow App")
    app.MainLoop()
```

```python
#38.Basic File Organizer
import wx
import os
import shutil

class FileOrganizerFrame(wx.Frame):
    def __init__(self, parent, title):
        super(FileOrganizerFrame, self).__init__(parent, title=title, size=(800, 600))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.BoxSizer(wx.VERTICAL)

        self.source_dir_text = wx.TextCtrl(self.panel, style=wx.TE_READONLY)
        self.browse_button = wx.Button(self.panel, label="Browse")
        self.organize_button = wx.Button(self.panel, label="Organize Files")

        self.browse_button.Bind(wx.EVT_BUTTON, self.OnBrowse)
        self.organize_button.Bind(wx.EVT_BUTTON, self.OnOrganize)

        self.sizer.Add(wx.StaticText(self.panel, label="Source Directory:"), flag=wx.LEFT | wx.TOP, border=10)
        self.sizer.Add(self.source_dir_text, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)
        self.sizer.Add(self.browse_button, flag=wx.ALIGN_RIGHT | wx.RIGHT | wx.BOTTOM, border=10)
        self.sizer.Add(self.organize_button, flag=wx.ALIGN_CENTER | wx.TOP | wx.BOTTOM, border=10)

        self.panel.SetSizer(self.sizer)

    def OnBrowse(self, event):
        dlg = wx.DirDialog(self, "Choose a directory to organize files from:", style=wx.DD_DEFAULT_STYLE)

        if dlg.ShowModal() == wx.ID_OK:
            source_dir = dlg.GetPath()
            self.source_dir_text.SetValue(source_dir)

        dlg.Destroy()

    def OnOrganize(self, event):
        source_dir = self.source_dir_text.GetValue()

        if not source_dir:
            wx.MessageBox("Please select a source directory.", "Error", wx.OK | wx.ICON_ERROR)
            return

        if not os.path.exists(source_dir):
            wx.MessageBox("The selected directory does not exist.", "Error", wx.OK | wx.ICON_ERROR)
            return

        file_types = set()

        for filename in os.listdir(source_dir):
            file_path = os.path.join(source_dir, filename)
            if os.path.isfile(file_path):
                file_extension = os.path.splitext(filename)[1][1:]
                if file_extension:
                    file_types.add(file_extension)

        if not file_types:
            wx.MessageBox("No files found in the selected directory.", "Information", wx.OK | wx.ICON_INFORMATION)
            return

        for file_type in file_types:
            target_dir = os.path.join(source_dir, file_type)
            os.makedirs(target_dir, exist_ok=True)
            for filename in os.listdir(source_dir):
                file_path = os.path.join(source_dir, filename)
                if os.path.isfile(file_path):
                    file_extension = os.path.splitext(filename)[1][1:]
                    if file_extension == file_type:
                        shutil.move(file_path, os.path.join(target_dir, filename))

        wx.MessageBox("Files have been organized successfully.", "Information", wx.OK | wx.ICON_INFORMATION)

if __name__ == '__main__':
    app = wx.App(False)
    frame = FileOrganizerFrame(None, "Basic File Organizer")
    app.MainLoop()
```

```python
#43.Basic Puzzle Game
import wx
import random

class PuzzleGameFrame(wx.Frame):
    def __init__(self, parent, title):
        super(PuzzleGameFrame, self).__init__(parent, title=title, size=(300, 300))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.GridSizer(3, 3, 0, 0)

        self.tiles = [str(i) for i in range(1, 9)] + [' ']
        random.shuffle(self.tiles)

        self.buttons = []

        for tile in self.tiles:
            button = wx.Button(self.panel, label=tile, size=(100, 100))
            button.Bind(wx.EVT_BUTTON, self.OnTileClick)
            self.buttons.append(button)
            self.sizer.Add(button, 0, wx.EXPAND)

        self.panel.SetSizerAndFit(self.sizer)

    def OnTileClick(self, event):
        button = event.GetEventObject()
        label = button.GetLabel()

        if label != ' ':
            row, col = self.GetTilePosition(label)
            if self.CanMoveTile(row, col):
                self.SwapTiles(row, col)
                self.UpdateButtonLabels()

                if self.IsPuzzleSolved():
                    wx.MessageBox("Congratulations! You solved the puzzle.", "Puzzle Solved", wx.OK | wx.ICON_INFORMATION)

    def GetTilePosition(self, tile_label):
        for row in range(3):
            for col in range(3):
                if self.tiles[row * 3 + col] == tile_label:
                    return row, col
        return -1, -1

    def CanMoveTile(self, row, col):
        empty_row, empty_col = self.GetTilePosition(' ')

        if abs(row - empty_row) + abs(col - empty_col) == 1:
            return True
        return False

    def SwapTiles(self, row, col):
        empty_row, empty_col = self.GetTilePosition(' ')
        index1 = row * 3 + col
        index2 = empty_row * 3 + empty_col
        self.tiles[index1], self.tiles[index2] = self.tiles[index2], self.tiles[index1]

    def UpdateButtonLabels(self):
        for button, tile_label in zip(self.buttons, self.tiles):
            button.SetLabel(tile_label)

    def IsPuzzleSolved(self):
        return self.tiles == ['1', '2', '3', '4', '5', '6', '7', '8', ' ']

if __name__ == '__main__':
    app = wx.App(False)
    frame = PuzzleGameFrame(None, "Basic Puzzle Game")
    app.MainLoop()
```

```python
#44.Basic Sudoku Game
import wx
import random

class SudokuGameFrame(wx.Frame):
    def __init__(self, parent, title):
        super(SudokuGameFrame, self).__init__(parent, title=title, size=(300, 350))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.GridSizer(9, 9, 0, 0)

        self.board = self.GenerateSudokuBoard()
        self.buttons = []

        for row in range(9):
            for col in range(9):
                cell_value = self.board[row][col]
                label = '' if cell_value == 0 else str(cell_value)
                button = wx.Button(self.panel, label=label, size=(30, 30))
                button.Bind(wx.EVT_BUTTON, self.OnCellClick)
                button.row = row
                button.col = col
                self.buttons.append(button)
                self.sizer.Add(button, 0, wx.EXPAND)

        self.panel.SetSizerAndFit(self.sizer)

    def GenerateSudokuBoard(self):
        # Generate a valid Sudoku board
        def solve(board):
            for row in range(9):
                for col in range(9):
                    if board[row][col] == 0:
                        for num in range(1, 10):
                            if self.IsValidMove(board, row, col, num):
                                board[row][col] = num
                                if solve(board):
                                    return True
                                board[row][col] = 0
                        return False
            return True

        board = [[0] * 9 for _ in range(9)]
        solve(board)
        self.RemoveNumbers(board, 40)  # Remove some numbers to create a puzzle
        return board

    def IsValidMove(self, board, row, col, num):
        # Check if placing num at (row, col) is a valid move
        return (
            self.IsRowValid(board, row, num) and
            self.IsColValid(board, col, num) and
            self.IsBoxValid(board, row - row % 3, col - col % 3, num)
        )

    def IsRowValid(self, board, row, num):
        return num not in board[row]

    def IsColValid(self, board, col, num):
        return num not in [board[row][col] for row in range(9)]

    def IsBoxValid(self, board, startRow, startCol, num):
        for row in range(3):
            for col in range(3):
                if board[row + startRow][col + startCol] == num:
                    return False
        return True

    def RemoveNumbers(self, board, num_to_remove):
        # Remove num_to_remove numbers from the Sudoku board
        for _ in range(num_to_remove):
            row, col = random.randint(0, 8), random.randint(0, 8)
            while board[row][col] == 0:
                row, col = random.randint(0, 8), random.randint(0, 8)
            board[row][col] = 0

    def OnCellClick(self, event):
        button = event.GetEventObject()
        row, col = button.row, button.col
        cell_value = self.board[row][col]

        if cell_value == 0:
            wx.MessageBox("You can't change this cell.", "Invalid Move", wx.OK | wx.ICON_ERROR)
        else:
            dlg = wx.TextEntryDialog(self, f"Enter a value for cell ({row + 1}, {col + 1}):", "Enter Value", str(cell_value))
            if dlg.ShowModal() == wx.ID_OK:
                input_value = dlg.GetValue()
                try:
                    input_value = int(input_value)
                    if 1 <= input_value <= 9:
                        if self.IsValidMove(self.board, row, col, input_value):
                            self.board[row][col] = input_value
                            button.SetLabel(str(input_value))
                        else:
                            wx.MessageBox("Invalid move. This number cannot be placed here.", "Invalid Move", wx.OK | wx.ICON_ERROR)
                    else:
                        wx.MessageBox("Please enter a number between 1 and 9.", "Invalid Input", wx.OK | wx.ICON_ERROR)
                except ValueError:
                    wx.MessageBox("Please enter a valid number between 1 and 9.", "Invalid Input", wx.OK | wx.ICON_ERROR)

if __name__ == '__main__':
    app = wx.App(False)
    frame = SudokuGameFrame(None, "Basic Sudoku Game")
    app.MainLoop()
```

```python
import wx
import math

class CalculatorFrame(wx.Frame):
    def __init__(self, parent, title):
        super(CalculatorFrame, self).__init__(parent, title=title, size=(300, 400))

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.GridBagSizer(5, 5)

        self.result_display = wx.TextCtrl(self.panel, style=wx.TE_RIGHT)
        self.result_display.SetValue("0")

        self.CreateButtons()

        self.sizer.Add(self.result_display, pos=(0, 0), span=(1, 4), flag=wx.EXPAND | wx.ALL, border=5)
        self.sizer.Add(self.button_sizer, pos=(1, 0), span=(1, 4), flag=wx.EXPAND | wx.ALL, border=5)

        self.panel.SetSizerAndFit(self.sizer)

    def CreateButtons(self):
        self.buttons = [
            '7', '8', '9', '/',
            '4', '5', '6', '*',
            '1', '2', '3', '-',
            '0', '.', '=', '+',
            'sqrt', 'sin', 'cos', 'tan',
            'log', 'exp', '(', ')',
            'AC', 'C',
        ]

        self.button_sizer = wx.GridSizer(6, 4, 5, 5)

        for label in self.buttons:
            button = wx.Button(self.panel, label=label, size=(50, 50))
            button.Bind(wx.EVT_BUTTON, self.OnButtonClick)
            self.button_sizer.Add(button, 0, wx.EXPAND)

    def OnButtonClick(self, event):
        button = event.GetEventObject()
        label = button.GetLabel()
        current_expression = self.result_display.GetValue()

        if label == '=':
            try:
                result = eval(current_expression)
                self.result_display.SetValue(str(result))
            except Exception as e:
                self.result_display.SetValue("Error")
        elif label == 'AC':
            self.result_display.SetValue("")
        elif label == 'C':
            current_expression = current_expression[:-1]
            self.result_display.SetValue(current_expression)
        elif label in ('sqrt', 'sin', 'cos', 'tan', 'log', 'exp'):
            function_mapping = {
                'sqrt': 'math.sqrt',
                'sin': 'math.sin',
                'cos': 'math.cos',
                'tan': 'math.tan',
                'log': 'math.log10',
                'exp': 'math.exp',
            }
            function_name = function_mapping[label]
            current_expression += function_name + '('
            self.result_display.SetValue(current_expression)
        else:
            current_expression += label
            self.result_display.SetValue(current_expression)

if __name__ == '__main__':
    app = wx.App(False)
    frame = CalculatorFrame(None, "Basic Calculator with Scientific Functions")
    app.MainLoop()
```

```python
#46.Basic Video Recorder
import wx
import os
import pyaudio
import wave
import threading

class VoiceRecorderFrame(wx.Frame):
    def __init__(self, parent, title):
        super(VoiceRecorderFrame, self).__init__(parent, title=title, size=(400, 200))

        self.InitUI()
        self.Centre()
        self.Show(True)

        self.p = pyaudio.PyAudio()
        self.frames = []

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.BoxSizer(wx.VERTICAL)

        self.record_button = wx.Button(self.panel, label="Record")
        self.stop_button = wx.Button(self.panel, label="Stop")
        self.play_button = wx.Button(self.panel, label="Play")
        self.save_button = wx.Button(self.panel, label="Save")

        self.record_button.Bind(wx.EVT_BUTTON, self.OnRecord)
        self.stop_button.Bind(wx.EVT_BUTTON, self.OnStop)
        self.play_button.Bind(wx.EVT_BUTTON, self.OnPlay)
        self.save_button.Bind(wx.EVT_BUTTON, self.OnSave)

        self.stop_button.Disable()
        self.play_button.Disable()
        self.save_button.Disable()

        self.sizer.Add(self.record_button, flag=wx.ALL, border=10)
        self.sizer.Add(self.stop_button, flag=wx.ALL, border=10)
        self.sizer.Add(self.play_button, flag=wx.ALL, border=10)
        self.sizer.Add(self.save_button, flag=wx.ALL, border=10)

        self.panel.SetSizer(self.sizer)

    def OnRecord(self, event):
        self.record_button.Disable()
        self.stop_button.Enable()
        self.play_button.Disable()
        self.save_button.Disable()

        self.frames = []
        self.stream = self.p.open(format=pyaudio.paInt16,
                channels=1,
                rate=44100,
                input=True,
                frames_per_buffer=1024)

        self.recording_thread = threading.Thread(target=self.RecordAudio)
        self.recording_thread.start()

    def RecordAudio(self):
        while True:
            data = self.stream.read(1024)
            self.frames.append(data)

    def OnStop(self, event):
        self.record_button.Enable()
        self.stop_button.Disable()
        self.play_button.Enable()
        self.save_button.Enable()

        self.stream.stop_stream()
        self.stream.close()

    def OnPlay(self, event):
        self.play_thread = threading.Thread(target=self.PlayAudio)
        self.play_thread.start()

    def PlayAudio(self):
        self.play_stream = self.p.open(format=self.p.get_format_from_width(2),
                channels=1,
                rate=44100,
                output=True)

        for frame in self.frames:
            self.play_stream.write(frame)

        self.play_stream.stop_stream()
        self.play_stream.close()

    def OnSave(self, event):
        save_dialog = wx.FileDialog(self, message="Save As", defaultDir=os.getcwd(), defaultFile="recording.wav",
            wildcard="Wave files (*.wav)|*.wav", style=wx.FD_SAVE)

        if save_dialog.ShowModal() == wx.ID_CANCEL:
            return

        file_path = save_dialog.GetPath()

        wf = wave.open(file_path, 'wb')
        wf.setnchannels(1)
        wf.setsampwidth(self.p.get_sample_size(pyaudio.paInt16))
        wf.setframerate(44100)
        wf.writeframes(b''.join(self.frames))
        wf.close()

if __name__ == '__main__':
    app = wx.App(False)
    frame = VoiceRecorderFrame(None, "Basic Voice Recorder")
    app.MainLoop()
```

```python
#47.Basic Video Recorder
import wx
import cv2

class VideoRecorderFrame(wx.Frame):
    def __init__(self, parent, title):
        super(VideoRecorderFrame, self).__init__(parent, title=title, size=(640, 480))

        self.InitUI()
        self.Centre()
        self.Show(True)

        self.cap = None
        self.is_recording = False
        self.video_writer = None

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.BoxSizer(wx.VERTICAL)

        self.start_button = wx.Button(self.panel, label="Start Recording")
        self.stop_button = wx.Button(self.panel, label="Stop Recording")

        self.start_button.Bind(wx.EVT_BUTTON, self.OnStartRecording)
        self.stop_button.Bind(wx.EVT_BUTTON, self.OnStopRecording)

        self.stop_button.Disable()

        self.sizer.Add(self.start_button, flag=wx.ALL, border=10)
        self.sizer.Add(self.stop_button, flag=wx.ALL, border=10)

        self.panel.SetSizer(self.sizer)

    def OnStartRecording(self, event):
        self.cap = cv2.VideoCapture(0)

        fourcc = cv2.VideoWriter_fourcc(*'XVID')
        self.video_writer = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

        self.is_recording = True
        self.start_button.Disable()
        self.stop_button.Enable()

        while self.is_recording:
            ret, frame = self.cap.read()
            if ret:
                self.video_writer.write(frame)

        self.cap.release()
        self.video_writer.release()

        self.start_button.Enable()
        self.stop_button.Disable()

    def OnStopRecording(self, event):
        self.is_recording = False

if __name__ == '__main__':
    app = wx.App(False)
    frame = VideoRecorderFrame(None, "Basic Video Recorder")
    app.MainLoop()
```

```python
#49.Basic Stopwatch App
import wx
import time

class StopwatchFrame(wx.Frame):
    def __init__(self, parent, title):
        super(StopwatchFrame, self).__init__(parent, title=title, size=(300, 150))

        self.InitUI()
        self.Centre()
        self.Show(True)

        self.is_running = False
        self.start_time = 0

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.BoxSizer(wx.VERTICAL)

        self.time_display = wx.StaticText(self.panel, label="00:00:00", style=wx.ALIGN_CENTER)
        font = wx.Font(24, wx.FONTFAMILY_DEFAULT, wx.FONTSTYLE_NORMAL, wx.FONTWEIGHT_NORMAL)
        self.time_display.SetFont(font)

        self.start_button = wx.Button(self.panel, label="Start")
        self.stop_button = wx.Button(self.panel, label="Stop")
        self.reset_button = wx.Button(self.panel, label="Reset")

        self.start_button.Bind(wx.EVT_BUTTON, self.OnStart)
        self.stop_button.Bind(wx.EVT_BUTTON, self.OnStop)
        self.reset_button.Bind(wx.EVT_BUTTON, self.OnReset)

        self.stop_button.Disable()
        self.reset_button.Disable()

        self.sizer.Add(self.time_display, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)
        self.sizer.Add(self.start_button, flag=wx.EXPAND | wx.ALL, border=5)
        self.sizer.Add(self.stop_button, flag=wx.EXPAND | wx.ALL, border=5)
        self.sizer.Add(self.reset_button, flag=wx.EXPAND | wx.ALL, border=5)

        self.panel.SetSizer(self.sizer)

        self.timer = wx.Timer(self)
        self.Bind(wx.EVT_TIMER, self.UpdateTime, self.timer)

    def OnStart(self, event):
        if not self.is_running:
            self.start_time = time.time() - self.start_time
            self.timer.Start(1000)
            self.is_running = True
            self.start_button.Disable()
            self.stop_button.Enable()
            self.reset_button.Disable()

    def OnStop(self, event):
        if self.is_running:
            self.timer.Stop()
            self.is_running = False
            self.start_button.Enable()
            self.stop_button.Disable()
            self.reset_button.Enable()

    def OnReset(self, event):
        self.start_time = 0
        self.is_running = False
        self.timer.Stop()
        self.start_button.Enable()
        self.stop_button.Disable()
        self.reset_button.Disable()
        self.UpdateTime()

    def UpdateTime(self, event=None):
        if self.is_running:
            elapsed_time = time.time() - self.start_time
            hours, remainder = divmod(int(elapsed_time), 3600)
            minutes, seconds = divmod(remainder, 60)
            self.time_display.SetLabel(f"{hours:02}:{minutes:02}:{seconds:02}")
        else:
            self.time_display.SetLabel("00:00:00")

if __name__ == '__main__':
    app = wx.App(False)
    frame = StopwatchFrame(None, "Basic Stopwatch App")
    app.MainLoop()
```

```python
#50.Basic Recipe Book
import wx

class RecipeBookFrame(wx.Frame):
    def __init__(self, parent, title):
        super(RecipeBookFrame, self).__init__(parent, title=title, size=(400, 300))

        self.recipes = []

        self.InitUI()
        self.Centre()
        self.Show(True)

    def InitUI(self):
        self.panel = wx.Panel(self)
        self.sizer = wx.BoxSizer(wx.VERTICAL)

        self.recipe_list = wx.ListBox(self.panel, choices=[], style=wx.LB_SINGLE)
        self.recipe_name_input = wx.TextCtrl(self.panel, placeholderText="Recipe Name")
        self.recipe_ingredients_input = wx.TextCtrl(self.panel, placeholderText="Ingredients", style=wx.TE_MULTILINE)
        self.recipe_instructions_input = wx.TextCtrl(self.panel, placeholderText="Instructions", style=wx.TE_MULTILINE)

        self.add_button = wx.Button(self.panel, label="Add Recipe")
        self.view_button = wx.Button(self.panel, label="View Recipe")
        self.edit_button = wx.Button(self.panel, label="Edit Recipe")
        self.delete_button = wx.Button(self.panel, label="Delete Recipe")

        self.recipe_name_input.Disable()
        self.recipe_ingredients_input.Disable()
        self.recipe_instructions_input.Disable()
        self.view_button.Disable()
        self.edit_button.Disable()
        self.delete_button.Disable()

        self.recipe_list.Bind(wx.EVT_LISTBOX, self.OnSelectRecipe)
        self.add_button.Bind(wx.EVT_BUTTON, self.OnAddRecipe)
        self.view_button.Bind(wx.EVT_BUTTON, self.OnViewRecipe)
        self.edit_button.Bind(wx.EVT_BUTTON, self.OnEditRecipe)
        self.delete_button.Bind(wx.EVT_BUTTON, self.OnDeleteRecipe)

        self.sizer.Add(self.recipe_list, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)
        self.sizer.Add(self.recipe_name_input, flag=wx.EXPAND | wx.ALL, border=10)
        self.sizer.Add(self.recipe_ingredients_input, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)
        self.sizer.Add(self.recipe_instructions_input, proportion=1, flag=wx.EXPAND | wx.ALL, border=10)
        self.sizer.Add(self.add_button, flag=wx.EXPAND | wx.ALL, border=5)
        self.sizer.Add(self.view_button, flag=wx.EXPAND | wx.ALL, border=5)
        self.sizer.Add(self.edit_button, flag=wx.EXPAND | wx.ALL, border=5)
        self.sizer.Add(self.delete_button, flag=wx.EXPAND | wx.ALL, border=5)

        self.panel.SetSizer(self.sizer)

    def OnSelectRecipe(self, event):
        selected_item = self.recipe_list.GetSelection()
        if selected_item != -1:
            recipe = self.recipes[selected_item]
            self.recipe_name_input.SetValue(recipe[0])
            self.recipe_ingredients_input.SetValue(recipe[1])
            self.recipe_instructions_input.SetValue(recipe[2])
            self.view_button.Enable()
            self.edit_button.Enable()
            self.delete_button.Enable()

    def OnAddRecipe(self, event):
        recipe_name = self.recipe_name_input.GetValue()
        recipe_ingredients = self.recipe_ingredients_input.GetValue()
        recipe_instructions = self.recipe_instructions_input.GetValue()

        if recipe_name and recipe_ingredients and recipe_instructions:
            self.recipes.append((recipe_name, recipe_ingredients, recipe_instructions))
            self.recipe_list.Append(recipe_name)
            self.ClearInputFields()

    def OnViewRecipe(self, event):
        pass  # You can implement this to display a dialog with the recipe details.

    def OnEditRecipe(self, event):
        selected_item = self.recipe_list.GetSelection()
        if selected_item != -1:
            recipe_name = self.recipe_name_input.GetValue()
            recipe_ingredients = self.recipe_ingredients_input.GetValue()
            recipe_instructions = self.recipe_instructions_input.GetValue()
            self.recipes[selected_item] = (recipe_name, recipe_ingredients, recipe_instructions)
            self.recipe_list.SetString(selected_item, recipe_name)

    def OnDeleteRecipe(self, event):
        selected_item = self.recipe_list.GetSelection()
        if selected_item != -1:
            del self.recipes[selected_item]
            self.recipe_list.Delete(selected_item)
            self.ClearInputFields()

    def ClearInputFields(self):
        self.recipe_name_input.Clear()
        self.recipe_ingredients_input.Clear()
        self.recipe_instructions_input.Clear()
        self.view_button.Disable()
        self.edit_button.Disable()
        self.delete_button.Disable()

if __name__ == '__main__':
    app = wx.App(False)
    frame = RecipeBookFrame(None, "Basic Recipe Book")
    app.MainLoop()
```
